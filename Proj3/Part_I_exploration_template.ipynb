## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your data. Make sure the variables that you cover here have been introduced in some fashion in the previous section (univariate exploration).


```python
# Scatter plot of Departure Delay vs Arrival Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='DepDelay', y='ArrDelay', alpha=0.5)
plt.title('Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.show()

# Box plot of Arrival Delays by Airline
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Reporting_Airline', y='ArrDelay', order=sorted(df['Reporting_Airline'].unique()))
plt.title('Arrival Delays by Airline')
plt.xlabel('Airline')
plt.ylabel('Arrival Delay (minutes)')
plt.xticks(rotation=90)
plt.show()

# Create a pivot table to summarize average arrival delays by airline and month
pivot_table = df.pivot_table(index='Reporting_Airline', columns='Month', values='ArrDelay', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Average Arrival Delay (minutes)'})
plt.title('Heatmap of Average Arrival Delays by Airline and Month')
plt.xlabel('Month')
plt.ylabel('Airline')
plt.show()

```

Some airlines exhibit higher median arrival delays and a wider spread of delay times, while others show more consistent performance with fewer extreme delays. This suggests operational differences among airlines that could be further explored.
>

Conversely, airlines KH exhibit more consistent performance across the year. These patterns suggest that both seasonal factors and airline-specific practices influence arrival delays.

## Multivariate Exploration

> Create plots of three or more variables to investigate your data even
further. Make sure that your investigations are justified, and follow from
your work in the previous sections.



```python
# Create a pivot table to summarize average arrival delays by airline and month
pivot_table = df.pivot_table(index='Reporting_Airline', columns='Month', values='ArrDelay', aggfunc='mean')

# Melt the pivot table to long format for seaborn FacetGrid
melted_df = pivot_table.reset_index().melt(id_vars='Reporting_Airline', var_name='Month', value_name='Average_ArrDelay')

# Ensure the 'Month' column is treated as a categorical variable
melted_df['Month'] = melted_df['Month'].astype(int)
melted_df['Month'] = pd.Categorical(melted_df['Month'], categories=sorted(melted_df['Month'].unique()), ordered=True)

# Initialize the FacetGrid
g = sns.FacetGrid(melted_df, col="Month", col_wrap=2, height=6, aspect=1.5)

# Function to create heatmaps within the FacetGrid
def heatmap(data, **kwargs):
    data = data.pivot(index='Reporting_Airline', columns='Month', values='Average_ArrDelay')
    sns.heatmap(data, **kwargs)

# Map the heatmap function to each subplot
g.map_dataframe(heatmap, cmap="YlGnBu", cbar_kws={'label': 'Average Arrival Delay (minutes)'}, annot=True, fmt=".1f")

# Adjust y-axis labels for readability
for ax in g.axes.flat:
    plt.setp(ax.get_yticklabels(), rotation=0)

g.set_titles("{col_name} Month")
g.set_axis_labels("Airline", "Average Arrival Delay (minutes)")
plt.show()



# Pairplot to see relationships between numerical features
sns.pairplot(df[['DepDelay', 'ArrDelay', 'AirTime', 'Distance']], diag_kind='kde')
plt.show()
```

### Talk about some of the relationships you observed in this part of the investigation. Were there features that strengthened each other in terms of looking at your feature(s) of interest?

> The heatmaps provide a clear view of how average arrival delays vary across airlines and months. This visualization highlights that certain airlines consistently have higher delays, suggesting persistent operational challenges or external factors impacting these airlines more than others. The pattern is more evident in the winter months, where delays are generally higher, likely due to adverse weather conditions.
> 
> The pair plot shows strong positive correlations between DepDelay (departure delay) and ArrDelay (arrival delay), reinforcing the idea that delays at departure strongly influence arrival times. This is further supported by the scatter plots showing a clear diagonal trend, indicating that as departure delays increase, arrival delays also increase proportionately. This relationship is crucial for understanding the propagation of delays through the flight schedule.

### Were there any interesting or surprising interactions between features?

> Yes, there were several interesting interactions observed. In the pair plot, AirTime and Distance show a strong positive correlation, which is expected since longer distances typically result in longer flight times. However, it's interesting to see the clustering in these plots, indicating groups of flights with similar characteristics.
> 
> Another surprising interaction is the presence of negative delay values in both departure and arrival delays, indicating early departures and arrivals. These instances can provide insights into operational efficiencies or favorable conditions that allow flights to arrive ahead of schedule.
>
